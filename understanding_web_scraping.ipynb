```python
# Import the requests, beautifulsoup4, time, pandas
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
from time import sleep

# Define the send_request_and_save_file() function
def send_request_and_save_file(url, filename):
    """
    Sends a request to the specified URL and saves the response content to the
    specified file.

    Args:
        url: The URL of the web page.
        filename: The filename of the file to save the request to.
    """

    # url to web scrape
    try:
        # Making a GET request to retrieve information from a server
        response = requests.get(url)

        if response.status_code == 200:
            # The request was successful, save the response content to the file
            with open(filename, "wb") as f:
                    f.write(response.content)
        else:
            # The request failed, print the status code
            print("Request failed with status code:", response.status_code)
    except Exception as e:
        print("An error occurred:", e)

# If this file is being executed as a script, run the send_request_and_save_file() function
if __name__ == "__main__":
    # The URL of the web page to request
    url = "https://inventwithpython.com/pythongently/"

    # The filename of the file to save the request to
    filename = "inventwithpython.html"

    # Send the request and save the response
    send_request_and_save_file(url, filename)
```

```python
def open_file(filename):
    """
    Opens the specified file and reads its contents.

    Args:
        filename: The filename of the file to open.

    Returns:
        The contents of the file as a string.
    """
    # Open the file in read mode
    with open(filename, "r") as f:
        contents = f.read()

    return contents

if __name__ == "__main__":
  # The filename of the file to open
  filename = "inventwithpython.html"

  # Open the file and print its contents
  contents = open_file(filename)
```

```python
# i need the <a>, href, text
my_content = focus_s.findAll('a')
```

```python
# create an empty list to push our final dict in it
final_content = []
```

```python
# we loop through through each <li> and get what we need
for license in my_content:
    my_dict = dict()
    title = license.text
    link = license.get('href')

    # calling the request function
    if __name__ == "__main__":
        # The URL of the web page to request
        child_link = 'https://inventwithpython.com/pythongently/' + link + '/'
        # The filename of the files to save the request to
        filenames = "inventwithpython" + ".html"

        # print(child_link)
        # Send the request and save the response
        link_body = send_request_and_save_file(child_link, filenames)

    # open the files
    body = open_file(filenames)


    link = child_link

    # use beautifulsoup to view and extract

    soup = bs(body, "html.parser")

    first_paragragh = soup.find(class_='Body').text

    # get only the first paragragh

    my_dict = {'title' : title, 'link': link, 'body': first_paragragh }
    # web-crawl : 2 sec
    time.sleep(1)
    print(my_dict)
    final_content.append(my_dict)
```
